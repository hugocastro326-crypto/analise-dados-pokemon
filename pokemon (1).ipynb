```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o dataset
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
df = pd.read_csv(url)

# Visualizando as primeiras linhas
df.head()

# Criando a coluna Total (soma dos principais atributos)
df["Total"] = df[["HP", "Attack", "Defense",
                  "Sp. Atk", "Sp. Def", "Speed"]].sum(axis=1)

# Verificando se a coluna foi criada corretamente
df[["Name", "Total"]].head()

# Ordenando do mais forte para o mais fraco
top_10 = df.sort_values("Total", ascending=False).head(10)

# Exibindo os 10 mais fortes
top_10[["Name", "Total", "Legendary"]]

# Calculando a média da força total por grupo
media_por_grupo = (
    df.groupby("Legendary")["Total"]
    .mean()
    .round(2)
    .reset_index()
)

media_por_grupo


plt.figure()

# Criando rótulos mais claros
labels = ["Não Lendário", "Lendário"]
valores = media_por_grupo["Total"]

# Criando gráfico
bars = plt.bar(labels, valores)

plt.title("Comparação da Média de Força")
plt.xlabel("Classificação")
plt.ylabel("Total Médio de Atributos")

# Adicionando valores acima das barras
for bar in bars:
    altura = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             altura,
             f"{altura:.2f}",
             ha='center',
             va='bottom')

plt.show()
```

# Análise de Dados do Universo Pokémon com Python

## Objetivo
Este projeto tem como objetivo analisar os atributos dos Pokémon para identificar padrões de força e verificar se Pokémon lendários possuem vantagem estatística em relação aos demais.

A análise foi realizada utilizando Python e a biblioteca Pandas.


## 1. Carregando os dados

Nesta etapa, carregamos o conjunto de dados contendo informações sobre os Pokémon, como HP, Attack, Defense, Speed, geração e classificação como lendário ou não.

O objetivo é preparar os dados para análise.


## 2. Criando a variável "Total"

Para facilitar a comparação entre os Pokémon, foi criada uma nova coluna chamada "Total", que representa a soma dos principais atributos:

HP + Attack + Defense + Sp. Atk + Sp. Def + Speed

Essa variável permite medir a força geral de cada Pokémon de maneira consolidada.


## 3. Identificando os Pokémon mais fortes

Nesta etapa, os Pokémon foram organizados do maior para o menor valor da variável "Total", permitindo identificar quais apresentam maior força geral.


## 4. Comparação entre lendários e não lendários

Nesta etapa, foi calculada a média da variável "Total" para dois grupos: Pokémon lendários e não lendários.

O objetivo é verificar se existe diferença significativa de poder entre esses dois grupos.


## Conclusão

A análise mostrou que:

- Pokémon lendários possuem, em média, maior valor total de atributos.
- A variável "Total" permite comparar a força geral de forma objetiva.
- A análise de dados possibilita identificar padrões que não são facilmente percebidos apenas pela observação direta.

Este projeto demonstra a aplicação de Python e Pandas para exploração e análise de dados estruturados.
